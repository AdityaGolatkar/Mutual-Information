```python
def plot_fisherORwtnorm(ax, experiments, title='Mutual Information', labels=None, hide_y = False, leg=False):
    y_base=20
    y_label=''
    legend_label=''
    
    exlog = experiments 

    key1=''
    key2=''
    
    #Select the legend depending on the phase of regularization
    if title=='early':legend_label='Until'
    elif title=='delay' or title=='slide':legend_label='From'
    elif title==None:print('Please select one of early/delay/slide for title to plot the legend')    
    
    #Select fisher vs wt_norm and appropriate y labels and x axis tick bases
    if fisher and not wt_norm:
        key1 = 'trace'
        key2 = 'trace'
        y_base=200
        y_label='Fisher Information Trace'
    elif not fisher and wt_norm:
        key1 = 'wt_norm'
        key2 = 'train'
        y_base=20
        y_label='$L_2$ norm of weights'
    elif fisher and wt_norm:
        print("Can't select fisher and wt_norm simultaneously")
    elif not fisher and not wt_norm: print('Please select at least one of fisher or wtnorm')
        
    #Choose colors
    color_idx = np.linspace(0, 1, 9)
    c_list = []
    for i in range(len(color_idx)):
        c_list.append(plt.cm.coolwarm(color_idx[i]))
        
    x, y = [], []
    for id_no,run_id in enumerate(exlog):

        try:
            logger = Logger.load(run_id)
        except ValueError:
            continue
        
        y = [v[key1] for v in logger.get(key2)]
        y = y[:200]
        if fisher and not wt_norm:
            y = savgol_filter(y,11,1)
        elif not fisher and wt_norm:
            if id_no!=0:
                y = [Logger.load(exlog[0]).get(key2)[0][key1]] + y
        x = np.arange(len(y))
        
        ax.xaxis.set_major_locator(tck.MultipleLocator(base=25))
        #if (wt_norm and title=='delay') or fisher:
        label = labels
        if label is None:
            label = '{0} {1}'.format(legend_label,id_no*25)
        #else:label = ''
        if not (title == 'early' and (id_no == len(exlog)-1)):
            ax.plot(x, y, label=label, color=c_list[id_no])

    ax.set_xlim(0,175)
    
    ax.yaxis.set_major_locator(tck.MultipleLocator(base=y_base))
    vals = ax.get_yticks()
    ax.set_yticklabels([int(x) for x in vals])

    ax.set_xlabel('Epoch')
    ax.set_ylabel(y_label)
    
    if title=='early':
        title='Early Termination of WD'
    elif title=='delay':
        title='Delayed Application of WD'
    elif title=='slide':
        title='Sliding Window (WD)'
    
    if title is not None:
        ax.set_title(title,fontsize=10)
    if hide_y:
        ax.yaxis.label.set_visible(False)
        
    if wt_norm and not fisher:
        if not leg:
#             if 'lid' in title or 'arly' in title:
            ax.legend(loc='center left',bbox_to_anchor=(l1,l2),fontsize=7,frameon=False,ncol=1,fancybox=True)
        else:
            ax.legend(loc='lower center',bbox_to_anchor=(0.5,1.15),fontsize=6,frameon=False,ncol=3,fancybox=True)
        
        #ax.legend(prop={'size': 7},bbox_to_anchor=(1.07, .5), loc='center left', borderaxespad=0.,frameon=False)
        #if 'elay' in title:
        #    ax.legend(loc='lower center',bbox_to_anchor=(1,1.1),fontsize=7,frameon=False,ncol=3,fancybox=True)
        #elif 'arly' in title:
        #ax.legend(loc='lower center',bbox_to_anchor=(0.5,1.1),fontsize=6,frameon=False,ncol=3,fancybox=True)
    elif not wt_norm and fisher:
        ax.legend(prop={'size': 7},frameon=False)
    
    return y
```
